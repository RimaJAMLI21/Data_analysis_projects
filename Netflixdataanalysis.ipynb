Q4. In which year highest number of the TV shows was released?
    show with Bar Graph.

```python
netflix
```

Extract the year

```python
netflix.Release_Date.unique()
```

```python
netflix['Release_Date'] = netflix['Release_Date'].str.lstrip()
```

```python
netflix['Release_Date'] = pd.to_datetime(netflix['Release_Date'],format='%B %d, %Y')
```

```python
netflix['Release_Date']
```

```python
netflix['Year'] = netflix['Release_Date'].dt.year
```

```python
netflix['Year'] = netflix['Year'].fillna(0)
```

```python
netflix['Year'].unique()
```

```python
netflix['Year'] = netflix['Year'].astype(int)
```

```python
netflix['Year'].unique()
```

```python
netflix[(netflix.Category == 'TV Show')].groupby('Year').size().sort_values(ascending = False)
 #netflix[netflix['Category'] == 'TV Show'].groupby('Year').size()
```

```python
netflix_bar_1 = netflix[(netflix.Category == 'TV Show')].groupby('Year').size().sort_values(ascending = False)
```

Bar Graph

```python
netflix_bar_1.plot(kind='bar')
plt.title("The TV shows released by Years")
```
